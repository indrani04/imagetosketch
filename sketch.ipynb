```python
import cv2
import numpy as np
```

```python
image = cv2.imread('megan.jpeg')

#converting into grey scale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#inverting the image
invert = cv2.bitwise_not(gray) #gray scale to inversion of the image

#smoothing the image
blurred = cv2.GaussianBlur(invert,(21,21),sigmaX=0,sigmaY=0)

#invert blurred image
invblur_img=cv2.bitwise_not(blurred)

sketch_img=cv2.divide(gray,invblur_img, scale=256.0)
```

```python
cv2.imshow("Final Photo",sketch_img)
cv2.imwrite("final_image.jpg",sketch_img)
cv2.waitKey(0) #wait s until the pressing any key.
cv2.destroyAllWindows() # this function destroy all windows of images after clicking any button.
```

```python
import matplotlib.pyplot as plt

img_color = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(img_color)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(sketch)
plt.axis('off')
plt.show()
```
